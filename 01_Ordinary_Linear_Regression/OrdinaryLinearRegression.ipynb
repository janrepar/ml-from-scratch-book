```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from LinearRegression import LinearRegression as LinearRegressionCustom
from sklearn.linear_model import LinearRegression
```

# Use of custom Linear regression function.

```python
df = pd.read_csv("../data/boston-housing/train.csv")
df.head()

X = df.drop(columns="medv").values
y = df["medv"].values
```

```python
model = LinearRegressionCustom()
model.fit(X, y, intercept=False)
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x=model.y, y=model.y_hat)
ax.set_xlabel(r'$y$', size = 16)
ax.set_ylabel(r'$\hat{y}$', rotation = 0, size = 16, labelpad = 15)
ax.set_title(r'$y$ vs. $\hat{y}$', size = 20, pad = 10)
sns.despine()
```

### Interpretation of the $y$ vs. $\hat{y}$ Graph

This **Predicted vs. Actual** plot helps evaluate the performance of the linear regression model.

*   **Axes**:
    *   The **x-axis ($y$)** shows the **actual values** (ground truth) from the dataset.
    *   The **y-axis ($\hat{y}$)** shows the **predicted values** generated by the model.

*   **Ideal Scenario**:
    *   In a perfect model, all points would lie on a **45-degree diagonal line** where $y = \hat{y}$.

*   **Analysis**:
    *   **Trend**: The upward diagonal trend indicates a strong positive correlation, meaning the model correctly predicts higher values for higher actual prices.
    *   **Residuals (Errors)**:
        *   Points **above** the diagonal represent **over-predictions** ($\hat{y} > y$).
        *   Points **below** the diagonal represent **under-predictions** ($\hat{y} < y$).
        *   The scatter (width of the cloud) represents the variance or error of the model; a tighter cloud implies higher accuracy.

# Scikit-Learn Implementation

```python
X_train = df.drop(columns="medv")
y_train = df["medv"]

sklearn_model = LinearRegression()
sklearn_model.fit(X_train, y_train)
```

```python
sklearn_predictions = sklearn_model.predict(X_train)

fig, ax = plt.subplots()
sns.scatterplot(x=y_train, y=sklearn_predictions)
ax.set_xlabel(r'$y$', size = 16)
ax.set_ylabel(r'$\hat{y}$', rotation = 0, size = 16, labelpad = 15)
ax.set_title(r'$y$ vs. $\hat{y}$', size = 20, pad = 10)
sns.despine()
```
